```python
# Import necessary machine learning models
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
# Prepare data for training
X = df.drop(columns=["species"])
Y = df["species"]

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.30)
```

```python
# Train and evaluate Logistic Regression model
model = LogisticRegression()
model.fit(x_train, y_train)
print("Logistic Regression Accuracy: ", model.score(x_test, y_test) * 100)
```

```python
# Train and evaluate K-nearest Neighbors model
model2 = KNeighborsClassifier()
model2.fit(x_train, y_train)
print("K-nearest Neighbors Accuracy: ", model2.score(x_test, y_test) * 100)
```

```python
# Train and evaluate Decision Tree model
model3 = DecisionTreeClassifier()
model3.fit(x_train, y_train)
print("Decision Tree Accuracy: ", model3.score(x_test, y_test) * 100)
```
